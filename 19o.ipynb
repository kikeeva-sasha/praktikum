```python
#игровое поле
maps=[1,2,3,4,5,6,7,8,9,
    1,1,1,2,1,3,1,4,1
    ,5,1,6,1,7,1,8,1,9]
#вывод поля на экран
def print_map():
    k=0
    for i in range(len(maps)):
        if k==8:
            print(maps[i])
            k=0
        else:
            print(maps[i],end=' ')
            k+=1


def hod(step1,step2,step3):
    if step1+step2!=10 and step1!=step2:
        print('Нет таких пар')
    for i in range(len(maps)+1):
        k=0
        if len(maps)-i>9:
            if step3=='g' and step1+step2==10 and maps[i]+maps[i+1]==10 and (maps[i]==step1 and maps[i+1]==step2) and i%8!=0:
                maps[i]=0
                maps[i+1]=0
                k+=1
                break
            elif step3=='g'and step1==step2 and maps[i]==maps[i+1] and maps[i]==step1 and maps[i+1]==step2 and i%8!=0:
                maps[i]=0
                maps[i+1]=0
                k+=1
                break
            elif step3=='v'and step1+step2==10 and maps[i]+maps[i+9]==10 and maps[i]==step1 and maps[i+9]==step2:
                maps[i] = 0
                maps[i +9] = 0
                k+=1
                break
            elif step3=='v' and step1==step2 and maps[i]==maps[i+9] and maps[i]==step1 and maps[i+9]==step2:
                maps[i]=0
                maps[i+9]=0
                k+=1
                break
        if len(maps) - i <= 9:
            if step3=='g' and step1+step2==10 and maps[i]+maps[i+1]==10 and maps[i]==step1 and maps[i+1]==step2:
                maps[i]=0
                maps[i+1]=0
                k+=1
                break
            elif step3=='g' and step1==step2 and maps[i]==maps[i+1] and maps[i]==step1 and maps[i+1]==step2:
                maps[i]=0
                maps[i+1]=0
                k+=1
                break
            else:
                break

    if k==0:
        for j in range(len(maps)):
            if maps[j] != 0:
                maps.append(maps[j])




    #новые числа(оставшиеся)




while True:
    print_map()
    para1=int(input("Выберите первое число: "))
    para2=int(input("Выберите второе число: "))
    kord=input('Выберите координату (v или g): ')
    if kord not in 'vg':
        print('Введена некорректная координата')
        continue

    hod(para1,para2,kord)
    if sum(maps)==0:
        print("Вы выиграли!")
        break
```
