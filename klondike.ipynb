```python
def draw_field(field):
    for row in field:
        print(' '.join(row))


def check_coordinates(x, y):
    if x < 0 or x >= len(field[0]):
        return False
    if y < 0 or y >= len(field):
        return False
    return True


def check_win(x, y):
    symbols = [field[y][x]]

    # Проверяем горизонталь и вертикаль
    for i in range(-3, 4):
        if i != 0 and check_coordinates(x + i, y) and field[y][x + i] != '-':
            symbols.append(field[y][x + i])
        if i != 0 and check_coordinates(x, y + i) and field[y + i][x] != '-':
            symbols.append(field[y + i][x])

    # Проверяем диагонали
    for i in range(-3, 4):
        if i != 0 and check_coordinates(x + i, y + i) and field[y + i][x + i] != '-':
            symbols.append(field[y + i][x + i])
        if i != 0 and check_coordinates(x - i, y + i) and field[y + i][x - i] != '-':
            symbols.append(field[y + i][x - i])

    # Проверяем, есть ли цепочка длиной хотя бы в 3 символа
    for symbol in symbols:
        if symbols.count(symbol) >= 3:
            return True

    return False


field = [['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-', '-', '-', '-', '-', '-']]

draw_field(field)

turn = 0
players = ['X', 'O']

while True:
    player = players[turn % 2]

    while True:
        x = int(input(f"Игрок {player}, введите номер столбца (1-10): "))
        y = int(input(f"Игрок {player}, введите номер строки (1-10): "))

        if not check_coordinates(x - 1, y - 1):
            print("Некорректные координаты! Попробуйте еще раз.")
            continue

        if field[y - 1][x - 1] != '-':
            print("Эта клетка уже занята! Попробуйте еще раз.")
            continue

        break

    field[y - 1][x - 1] = player
    draw_field(field)

    if check_win(x, y):
        print(f"Игрок {player} проиграл!")
        break

    turn += 1
    if turn == len(field) * len(field[0]):
        print("Игра окончена. Ничья!")
        break
```
