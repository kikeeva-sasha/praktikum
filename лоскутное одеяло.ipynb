```python
def draw_field(field):
    for row in field:
        print(' '.join(row))


def check_coordinates(x, y):
    if x < 0 or x >= len(field[0]):
        return False
    if y < 0 or y >= len(field):
        return False
    return True


def calculate_penalty(x, y):
    penalty = 0
    symbol = field[y][x]

    # Проверяем соседние клетки по горизонтали и вертикали
    if x > 0 and field[y][x - 1] == symbol:
        penalty += 1
    if x < len(field[0]) - 1 and field[y][x + 1] == symbol:
        penalty += 1
    if y > 0 and field[y - 1][x] == symbol:
        penalty += 1
    if y < len(field) - 1 and field[y + 1][x] == symbol:
        penalty += 1

    # Проверяем соседние клетки по диагонали
    if x > 0 and y > 0 and field[y - 1][x - 1] == symbol:
        penalty += 1
    if x < len(field[0]) - 1 and y > 0 and field[y - 1][x + 1] == symbol:
        penalty += 1
    if x > 0 and y < len(field) - 1 and field[y + 1][x - 1] == symbol:
        penalty += 1
    if x < len(field[0]) - 1 and y < len(field) - 1 and field[y + 1][x + 1] == symbol:
        penalty += 1

    return penalty


field = [['-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-']]

players = ['X', 'O', '#']
penalties = [0, 0, 0]

draw_field(field)

for turn in range(len(field[0]) * len(field)):
    player = players[turn % len(players)]

    while True:
        x = int(input(f"Игрок {player}, введите номер столбца (1-5): "))
        y = int(input(f"Игрок {player}, введите номер строки (1-4): "))

        if not check_coordinates(x - 1, y - 1):
            print("Некорректные координаты! Попробуйте еще раз.")
            continue

        if field[y - 1][x - 1] != '-':
            print("Эта клетка уже занята! Попробуйте еще раз.")
            continue

        break

    field[y - 1][x - 1] = player
    penalties[turn % len(players)] += calculate_penalty(x - 1, y - 1)
    draw_field(field)

min_penalty = min(penalties)
winners = [players[i] for i, penalty in enumerate(penalties) if penalty == min_penalty]

print("Игра окончена!")
print(f"Победители: {' '.join(winners)}")
print(f"Штрафные очки: {penalties}")
```
